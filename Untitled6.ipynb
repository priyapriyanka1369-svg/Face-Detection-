```python
from matplotlib import pyplot as pyplot
image = cv2.imread(image_file)
if image is None:
    print(f"Error: Could not load image from {image_file}")
else:
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pyplot.imshow(image)
    pyplot.show()
```

```python
!wget -O haar_face.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier('haar_face.xml')
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor =1.1, minNeighbors = 5)
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(image_rgb, (x,y), (x+w, y+h), (255,0,0), 2)
```

```python
pyplot.figure(figsize=(8,6))
pyplot.imshow(image_rgb)
pyplot.title(f'Faces detected: {len(faces)}')
pyplot.axis('off')
pyplot.show()
```

```python
import cv2
from matplotlib import pyplot as pyplot

image_file = "/content/R.jpg"
image = cv2.imread(image_file)
if image is None:
    print(f"Error: Could not load image from {image_file}")
else:
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pyplot.imshow(image_rgb)
    pyplot.show()
```

```python
conversation_data = {
    "hello":"Hi there! How can I help you today",
    "hi":"Hello! How are you?",
    "how are you":"I'm just a bot, but I'm doing great! How about you?",
    "what is computer vision":"Computer vision is a field of AI that allows computers to see and interpret images and video.",
    "what is ai":"AI stands for Artificial Intelligence, which enables machines to perform tasks that normally require human intelligence.",
    "what is deep learning":"Deep Learning is a subset of AI that uses neural networks to learn from large amounts of data.",
    "bye":"Goodbye! Have a nice day!",
    "thank you":"You're welcome! Happy to help."
}
```

```python
def sample_chat(user_input, chat_history=[]):
  user_input_lower = user_input.lower()
  response = None

  for key in conversation_data:
    if key in user_input_lower:
      response = conversation_data[key]
      break

  if not response:
    response = "Sorry, I don't understand that. Can you ask something else?"

  chat_history.append(("You", user_input))
  chat_history.append(("Bot", response))

  return response, chat_history
```

```python
print("Welcome to Mini Conversation AI!")
print("Type 'exit' or 'quit' to end the chat.\n")

chat_history = []

while True:
  User_input = input("you:")

  if User_input.lower() in ["exit", "quit"]:
    print("chat ended: Goodbye!")
    break

  bot_response, chat_history = sample_chat(User_input, chat_history)
  print("bot:", bot_response)
```

```python
print("\n=== full conversation history ===")
for speaker,message in chat_history:
  print(f"{speaker}: {message}")
```
